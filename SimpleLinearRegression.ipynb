```python
#normalizaci√≥n
x_media = np.concatenate((train_ex1,test_ex1),axis=0).mean(axis=0)
x_std = np.concatenate((train_ex1,test_ex1),axis=0).std(axis=0)
y_media = np.concatenate((y_train_ex1data1,y_test_ex1),axis=0).mean()
y_std = np.concatenate((y_train_ex1data1,y_test_ex1),axis=0).std()
print(x_media)
print(x_std)
print(y_media)
print(y_std)
```

```python
x_train = (train_ex1 - x_media)/x_std
x_test = (test_ex1 - x_media)/x_std
y_train = (y_train_ex1data1 - y_media)/y_std
y_test= (y_test_ex1-y_media)/y_std

print(x_train)
print(x_test)
print(y_train)
print(y_test)
```

```python
plt.plot(train_ex1,'o')
plt.plot(x_train,'o',color="black")
```

```python
n_exa_train = len(y_train)
n_exa_test = len(y_test)

x_train = np.concatenate((np.ones([n_exa_train,1]),x_train),axis=1)
x_test = np.concatenate((np.ones([n_exa_test,1]),x_test),axis=1)

print(x_train)
plt.plot(x_train,'o',color="black")
```
